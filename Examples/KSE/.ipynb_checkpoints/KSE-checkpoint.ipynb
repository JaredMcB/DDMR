```julia

Δt = h*obs_gap

function InvBurgRK4_1step(x)
  lx = length(x)
  function F(x)
      𝑥 = [conj(reverse(x, dims = 1));0; x]
      -1/2*(2π/P*(1:lx)/N) .* conv(𝑥,𝑥)[2*lx+2:3*lx+1]
  end

  k1 = F(x)
  k2 = F(x .+ Δt*k1/2)
  k3 = F(x .+ Δt*k2/2)
  k4 = F(x .+ Δt*k3)
  A =  @. x + Δt/6*(k1 + 2k2 + 2k3 + k4)
end

function Inertialman_part_mat(x)
  lx = length(x)
  𝑥(j) = ( j <= lx ? x[j] : im*sum(x[l]*x[j-l] for l = j-lx:lx) )

  L = complex(zeros(lx,lx))
  for j = 1:lx
     for k = 1:lx
        L[k,j] = 𝑥(j+lx)*𝑥(j+lx-k)
     end
  end
  L
end

function Inertialman_part(x)
  lx = length(x)
  𝑥(j) = ( j <= lx ? x[j] : im*sum(x[l]*x[j-l] for l = j-lx:lx) )

  L = complex(zeros(lx^2))
  for j = 1:lx
     for k = 1:lx
        L[ (j-1)*lx+k] = 𝑥(j+lx)*𝑥(j+lx-k)
     end
  end
  L
end


Psi(x) = [x; InvBurgRK4_1step(x); Inertialman_part(x)]
Psi_mat(x) = [x InvBurgRK4_1step(x) Inertialman_part_mat(x)]
```
