# Model Reduction on a Linear SDE

```julia
include("modgen_LSDE.jl")
include("..\\..\\Tools\\Model_Reduction_Dev.jl")
```

```julia
using JLD
using PyPlot
using DSP: nextfastfft
```

## Generate the data 

```julia
A = -[0.1 10; 0 0.9]
σ = [1 0; 0 1]
Xo = [1; 1]
t_disc = 1000
gap = 1
scheme = "EM"

t_start = 0
t_stop = 5e3
h = 0.01

X = modgen_LSDE(t_start,t_stop,h,
    A = A,
    σ = σ,
    Xo = Xo,
    t_disc = t_disc,
    gap = gap)
```

```julia
plot(X[1,:],X[2,:])
```

```julia
Psi(x) = x

@time h_wf = get_wf(X,Psi, M_out = 20)
```

```julia
I + h*A
```

## Get reduced (reproduced) model (no noise) 

```julia
d, N  = size(X)
nu    = size(Psi(X[:,1]),1) 
M_out = size(h_wf,3)

X_rm = zeros(d,N); X_rm[:,1:M_out] = X[:,1:M_out]

PSI = zeros(nu,N); 
for i = 1:M_out
    PSI[:,i] = Psi(X_rm[:,i])
end

for i = M_out + 1 : N
    X_rm[:,i] = sum(h_wf[:,:,k]*PSI[:,i-k] for k = 1:M_out, dims = 2)
    PSI[:,i] = Psi(X_rm[:,i])
end

```

```julia
data = Dict(
        "h_wf_4" => h_wf_4,
        "h_wf_5" => h_wf_5,
        "h_wf_55"=> h_wf_55,
        "A" => A,
        "σ" => σ,
        "Xo" => Xo,
        "t_disc" => t_disc,
        "gap" => gap,
        "scheme" => scheme,
        "t_start" => t_start,
        "t_stop" => t_start,
        "h" => h,
        "X_55" => X,
        "X__rm_55_h_4" => X_rm)

save("Data\\LSDE_wfs.jld",data)
```
