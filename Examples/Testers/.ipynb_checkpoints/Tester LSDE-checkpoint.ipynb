# Tester: Linear SDE's

A Linear SDE for a $n$-dimensional system is of the form
    $$ dX_t = AX_tdt + \sigma dW_t $$
where $W_t$ is a $m$-dimensional Wiener process.

```julia
using PyPlot
using Random

# Get software to generate model
include("../LinearSDE/modgen_LSDE.jl")

# Get model reduction software being tested
include("../../Tools/Model_Reduction_Dev.jl")

# Get tester tools
include("testertools.jl")
```

## 1-D 

```julia
# Model run Parameters
t_start = 0
t_stop  = 1e3
h       = 1e-2

A       = reshape([-0.5],1,1)
σ       = reshape([1],1,1)
Xo      = [1]
t_disc  = 100
gap     = 1

# Get full model run
Random.seed!(2014)
X = modgen_LSDE(t_start,t_stop,h;
    A, σ, Xo, t_disc, gap)
```

```julia
# Put in Psi functions
Psi(x) = x
```

```julia
# Model reduction Parameters
M_out = 100
n = 2
p = 500
par = 55
ty = "bin"
rl = true
Preds = true
PI = false
rtol = 1e-6

nfft = 1024

@time h_wf, pred = get_wf(X, Psi;
    M_out, n, p, par, ty, nfft, rl, Preds, PI, rtol);

X_sig = X[:,2:end]
```

```julia
pred
```

## Analyze Wiener Filter

```julia
h_wf
```

```julia
plot_WF(h_wf, rl = false)
```

### Poles of Wiener Filter

## Analyze One step prediction error

```julia
X_hat = one_step_pred(X_sig, h_wf, pred)
```

```julia
wind = (1:100) .+ 24000
plot([X_sig[1,wind] X_hat[1,wind]])
```

```julia
X_err = X_sig - X_hat

C = my_crosscov(pred[:],X_err[:],-100:10)

plot(C)
```

## Analyze Reduced Model Run

```julia
X_rm = redmodrun(X, h_wf, Psi)
```

## 2-D 

```julia
# Model run Parameters
t_start = 0
t_stop  = 1e3
h       = 1e-2

A       = [-0.5 0; 0 -0.2]
σ       = [1 0; 0 1]
Xo      = [1; 1]
t_disc  = 100
gap     = 1

# Get full model run
Random.seed!(2014)
Y = modgen_LSDE(t_start,t_stop,h;
    A, σ, Xo, t_disc, gap)
```

```julia
# Put in Psi functions
Psi(x) = x
```

```julia
# Model reduction Parameters
M_out = 100
n = 2
p = 500
par = 55
ty = "bin"
rl = true
Preds = true
PI = false
rtol = 1e-6

nfft = 1024

@time h_wf, pred = get_wf(Y, Psi;
    M_out, n, p, par, ty, nfft, rl, Preds, PI, rtol);

Y_sig = Y[:,2:end]
```

```julia
h_wf
```

```julia
pred
```

```julia
Y_hat = one_step_pred(Y_sig, h_wf, pred)
```

```julia
wind = (1:100) .+ 24000
plot([Y_sig[1,wind] Y_hat[1,wind] Y_sig[2,wind] Y_hat[2,wind]])
```

```julia
Y_err = Y_sig - Y_hat

C1 = my_crosscov(pred[1,:],Y_err[1,:],-100:10)
C2 = my_crosscov(pred[2,:],Y_err[2,:],-100:10)

plot([C1 C2])
```

```julia
plot(Y[:,1:10:end]')
```

```julia
noise_dist = MvNormal(h*I + zeros(size(Y,1),size(Y,1)))
```

```julia
Y_rm = redmodrun(Y, h_wf, Psi;noise_dist)
```

```julia
Y_rm
```

```julia
plot(Y_rm[:,1:10:end]')
```

```julia
C = matrix_autocov_seq(Y;
    L = 500, win = "par"
    )

C_rm = matrix_autocov_seq(Y_rm;
    L = 500, win = "par"
    )
```

```julia
plot_WF(C)
plot_WF(C_rm)
```
