## Producing the data

For this notebook we consider the stochastic, overdamped Langevin equation with a double-welled potential $V(x) = \frac{1}{4}(x^2 - 1)^2$. So, we consider the following SDE. 
$$dX_t = -X_t(X_t^2 - 1)dt + dB_t$$
To find a numerical solution we first use the Euler-Maruyama scheme 
$$X_{n+1} = X_n - h X_n(X_n^2 - 1) + e_n$$
here $$e_n \sim (B_{t_{n+1}}-B_{t_{n}}) \sim N(0,h) \sim \sqrt{h} N(0,1).$$

Now we generate the sample path.

```julia
# Packages used
using Plots
pyplot()

include("wiener_filter_Matrix_fft.jl")
```

```julia
t_start = 0
t_stop = 10^3
steps = 10^5 + 1 # (this includes the t_stop)
discard = 10^4

steps_tot = steps + discard

dVdx(x) = -x*(x^2 -1) 

sig_init = 1.5

h = (t_stop - t_start)/(steps)
t = range(t_start,stop = t_stop,length = steps)

e = randn(steps_tot)
signal = zeros(steps_tot)
signal[1] = sig_init

for n = 2 : steps_tot
    signal[n] = signal[n-1] + h*dVdx(signal[n-1]) + sqrt(h)*e[n]
end

sig = signal[discard+1:steps_tot]
plot(t,sig)
```

We seek to fit a NARMAX representation of the signal the exogenous terms will be $\Psi(x) = (\psi_1(x),\psi_2(x))^T) := (x, V'(x))$. To do this we will use a Wiener filter with predictors $\Psi_n = \Psi(Y_n)$. 

```julia
Psi(x) = [x; x^3]

sig = reshape(sig,1,steps)

pred = zeros(2, steps)
for n = 1:steps 
    pred[:,n] = Psi(signal[discard + n - 1])
end

M_h = 20
h_wf = vector_wiener_filter_fft(pred, sig, M_h, par = 60, Nex = 2^10, win = "Par")
h_wf = real.(h_wf)
```

```julia
plot(1:20, abs.([h_wf[1,1,:] h_wf[1,2,:]]))
```

```julia
sig_hat = zeros(1,steps)
for i = 1:steps
    sig_hat[:,i] = sum([real.(h_wf[:,:,k+1])*pred[:,i-k] for k = 0:min(i - 1,M_h - 1)])
end
```

```julia
view = 50
start = steps - 10 - 50 
tim = start:start + view;

plot(tim,[sig[1,tim] sig_hat[1,tim]],
    color=[:red :black], 
    line=(2,[:solid :dash]),
    label=["sig" "sig_hat" ])
```

```julia
error = sig[:] .- sig_hat[:]



lags = -200:200
C3 = crosscov(error,pred[1,:],lags)
C4 = crosscov(error,pred[2,:],lags)

plot(lags, [C1 C2 C3 C4],
    color=[:black :red],
    line=(2,[:dash :dot ]),
    label=["Error pred1 old" "Error pred2 old" "Error pred1 new" "Error pred2 new"],
    marker=([:d],3))
```

```julia
C1, C2 = C3, C4;
```
