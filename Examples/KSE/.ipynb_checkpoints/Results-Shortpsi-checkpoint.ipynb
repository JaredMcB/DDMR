```julia
# Load and look at original sol

Dat_sol = load("Data\\KSE_sol2.jld")
# here are the run parameters
```

```julia
uu = Dat_sol["dat_uu"]
vv = Dat_sol["dat_vv"]
h = Dat_sol["h"]
H1 = heatmap(uu[:,1:100:Int(300÷h)]')
```

```julia
Dat_wf = load("Data\\KSE_sol_wienerfilter_short$gen.jld")
h_wf = Dat_wf["dat_h_wf"]
```

```julia
d, nu, M_out = size(h_wf)

H_wf = zeros(d*M_out,nu)
for i = 1:M_out
    H_wf[d*(i-1)+1:d*i,:] = h_wf[:,:,i]
end

H_H_wf = heatmap(log.(10,abs.(H_wf[1:d*M_out,:])),
    title = "heatmap of log abs of stacked h_wf gen$gen",
    xaxis = "coefficients to Psi",
    yaxis = "(for h_wf[i,:,k]) row = 5(k-1) + i")
```

```julia
savefig(H_H_wf,"H_wfgen_short$gen.png")
```

```julia
plot(log.(10,abs.(H_wf[1:5:end,1:25])),leg = :none)
```

## Run the reduced model

```julia
Dat_rm = load("Data\\KSE_sol_redmodrun_short$gen.jld")
```

```julia
uu_rm = Dat_rm["dat_UU_rm"]
vv_rm = Dat_rm["dat_VV_rm"]
```

```julia
blup = findall(isnan,uu_rm[1,:])[1]
println("Blow-up step $blup")
```

```julia
cldn = findall(x -> x < 1e-6,sum(abs.(uu_rm),dims = 1)[:])[1]    
```

```julia
H1 = heatmap(uu_rm[:,1:1500]')
```

```julia
v2 = real(vv[2,1:100:150000])
v_rm2 = real(vv_rm[2,1:1500])

A2 = autocor(v2,0:150)
A_rm2 = autocor(v_rm2,0:150)

v3 = real(vv[3,1:100:150000])
v_rm3 = real(vv_rm[3,1:1500])

A3 = autocor(v3,0:150)
A_rm3 = autocor(v_rm3,0:150)
```

```julia
plot([A2 A_rm2 A3 A_rm3], line = (3,[:solid :dot :solid :dot]),
    color = [:red :red :green :green])
```
