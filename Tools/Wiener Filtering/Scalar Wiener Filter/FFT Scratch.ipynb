```julia
r, t = .9, pi/2

# each need modulus less than one for the process to be stationary. 
# Also for the process to be rea-valued r1 and r2 are either real or complex conjugates
r1, r2 = r*exp(im*t), r*exp(-im*t) 
sig_v = 1
M_h = 50
par = 55

sig_v = 10

l = [1, -(r1 + r2), r1*r2];

steps = 1*10^6
discard = 10^3
steps_tot = steps + discard

noise_sig = randn(steps_tot)
sig = zeros(steps_tot);

sig[1:2] = noise_sig[1:2];
for i = 3 : steps_tot
    sig[i] = dot(-reverse(l)[1:2],sig[i - 2:i-1]) + noise_sig[i]
end
sig = sig[discard + 1 : steps_tot];

noise_pred = sig_v*randn(steps)
pred = sig .+ noise_pred;
```

### Study of Crosscovariance

```julia
lags = -60:60;
C = crosscov(sig,pred,lags);
```

```julia
function my_crosscov(x,y,lags)
    L = min(length(x),length(y))
    m = length(lags)
    
    zx = x .- mean(x)
    zy = y .- mean(y)
    
    r = zeros(m)
    for k = 1:m
        l = lags[k]
        r[k] = ( l >= 0 ? dot(zx[1+l : L],zy[1 : L-l]) : dot(zx[1 : L+l],zy[1-l : L]))/L
    end
    r
end

function their_crosscov(x,y,lags)
    lx = length(x)
    m = length(lags)
    
    zx = x .- mean(x)
    zy = y .- mean(y)
    
    r = zeros(m)
    for k = 1:m
        l = lags[k]
        r[k] = ( l >= 0 ? dot(zx[1 : lx-l],zy[1+l : lx]) : dot(zx[1-l : lx],zy[1 : lx+l]))/lx
    end
    r
end
```

```julia
lags = -60:60;

C = crosscov(sig,pred,lags);
c = my_crosscov(sig,pred,lags);
C_ana = [r1^abs(s) + r2^abs(s) for s = -60:60]*C[61]/2;
```

###  Study of z-cross-spectra

```julia
function z_crossspect(sig,pred,L; win = "Bart")
    lags = -L:L;
    C = crosscov(sig,pred,lags)
    if win == "Bar"
        lam = 1 .- (0:L)/L
    elseif win == "Tuk"
        lam = .5*(1 .+ cos.(pi/L*(0:L)))
    elseif win == "Par"
        LL = Int(floor(L/2))
        lam1 = 1 .- 6*((0:LL)/L).^2 .+ 6*((0:LL)/L).^3
        lam2 = 2*(1 .- (LL+1:L)/L).^3
        lam = [lam1; lam2]
    else
        lam = ones(L+1)
    end

    z_crossspect_num(z) = sum([lam[abs(i) + 1]*C[L+1+i]*z^(-i) for i = -L : L])
end
```

```julia
function z_crossspect_fft(sig,pred,L,Nex; win = "Par")
    Nexh = Int(floor(Nex/2))
    lags = -L:L;
    c = crosscov(sig,pred,lags);

    # Smoothing window
    if win == "Bar"
        lam = 1 .- (0:L)/L
    elseif win == "Tuk"
        lam = .5*(1 .+ cos.(pi/L*(0:L)))
    elseif win == "Par"
        LL = Int(floor(L/2))
        lam1 = 1 .- 6*((0:LL)/L).^2 .+ 6*((0:LL)/L).^3
        lam2 = 2*(1 .- (LL+1:L)/L).^3
        lam = [lam1; lam2]
    else
        lam = ones(L+1)
    endFF
    Lam = [lam[L+1:-1:2]; lam]

    C_smoothed = Lam .* c;
    C_padded = [zeros(Nex - Nexh - L); C_smoothed; zeros(Nexh - L - 1)];
    C = fftshift(C_padded);

    z_crossspect_num_fft = fft(C)
end
```

$$\operatorname{DFT}(A)[k] =
  \sum_{n=1}^{\operatorname{length}(A)}
  \exp\left(-i\frac{2\pi
  (n-1)(k-1)}{\operatorname{length}(A)} \right) A[n].$$

```julia
z_crossspect_sigpred_num = z_crossspect(sig, pred, par, win = "Par");

z_crossspect_sigpred_num_fft = z_crossspect_fft(sig, pred, 120, Nex, win = "Par");

z_crossspect_sigpred_ana(z) = ( (1-r1*z^(-1))*(1-r2*z^(-1))*(1-conj(r1)*z)*(1-conj(r2)*z) )^(-1);
```

```julia
view = 200:824
plot(Theta,([real.(z_crossspect_sigpred_ana.(Z)
                ) (real.(z_crossspect_sigpred_num.(Z))
                ) (real.(z_crossspect_sigpred_num_fft))]),
    color = [:red :black :gray],
    line = ([:solid :dash :dot],2))
```

### Simple function example

The example here if $f(x) = 3 + z^{-1} + z$.

```julia
c = [zeros(511); 1; 3; 1; zeros(510)];
f(z) = 3 + z^(-1) + z;
F(th) = 3 + 2*cos(th);

f_fft = fft(fftshift(c));

Nex = 2^10;
Theta = 2pi*(0:Nex-1)/(Nex-1);
Z = map(th -> exp(im*th),Theta);

plot(Theta,([real.(f.(Z)
                ) real.(F.(Theta)
                ) real.(f_fft)]),
    color = [:red :black :gray],
    line = ([:solid :dash :dot],2))
```

The example here if $f(x) = \dfrac{1}{3 + z^{-1} + z}$.

```julia
Nexh = 2^9

z_hat = (-3 + sqrt(5))/2;

c = [sum([-z_hat^(2*k+abs(n)+1) for k = 0:100]) for n = -Nexh:Nexh - 1]

f(z) = (3 + z^(-1) + z)^(-1);
F(th) = 1/(3 + 2*cos(th));

f_fft = fft(fftshift(c));

Nex = 2^10;
Theta = 2pi*(0:Nex-1)/(Nex-1);
Z = map(th -> exp(im*th),Theta);

plot(Theta,([real.(f.(Z)
                ) real.(F.(Theta)
                ) real.(f_fft)]),
    color = [:red :black :gray],
    line = ([:solid :dash :dot],2))
```
