```julia
using PyPlot
using StatsBase


include("modgen_LSDE.jl")
```

```julia
A = reshape([-0.05],1,1)
σ = reshape([1],1,1)
Xo = [1]
t_disc = 1000
gap = 1
scheme = "EM"

X = modgen_LSDE(0,10000,0.01,
    A = A,
    σ = σ,
    Xo = Xo,
    t_disc = t_disc,
    gap = gap,
    scheme = scheme)
```

```julia
plot(X[1,1:200:end])
```

```julia
A1 = autocor(X[1,:],1:4000)
plot(A1)
```
