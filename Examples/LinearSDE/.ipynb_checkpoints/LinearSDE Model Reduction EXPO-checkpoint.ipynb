# Model Reduction on a Linear SDE

```julia
include("modgen_LSDE.jl")
include("..\\..\\Tools\\Model_Reduction_Dev.jl")
```

```julia
using JLD
using PyPlot
using DSP: nextfastfft
```

## Generate the data 

```julia
A = -[0.5 0; 0 0.9]
σ = [1 0; 0 1]
Xo = [1; 1]
t_disc = 1000
gap = 1
scheme = "EM"

t_start = 0
t_stop = 5e3
h = 0.01

X = modgen_LSDE(t_start,t_stop,h,
    A = A,
    σ = σ,
    Xo = Xo,
    t_disc = t_disc,
    gap = gap)
```

```julia
plot(X[1,:],X[2,:])
```

```julia
Psi(x) = x
n = 2
p = 100
ty = "bin"
@time h_wf = get_wf(X,Psi;n,p,ty,M_out = 20)
```

```julia
h_ana = zeros(2,2,1)
h_ana[:,:,1] =I + h*A
```

## Get reduced (reproduced) model (no noise) 

```julia
# h_wf = h_ana

d, N  = size(X)
nu    = size(Psi(X[:,1]),1) 
M_out = size(h_wf,3)

X_rm = zeros(d,N); X_rm[:,1:M_out] = X[:,1:M_out]

PSI = zeros(nu,N); 
for i = 1:M_out
    PSI[:,i] = Psi(X_rm[:,i])
end

for i = M_out + 1 : N
    X_rm[:,i] = sum(h_wf[:,:,k]*PSI[:,i-k] for k = 1:M_out, dims = 2)
    PSI[:,i] = Psi(X_rm[:,i])
end

```

```julia
d, N  = size(X)
nu    = size(Psi(X[:,1]),1) 
M_out = size(h_wf,3)

X_rm_n = zeros(d,N); X_rm_n[:,1:M_out] = X[:,1:M_out]

PSI_n = zeros(nu,N); 
for i = 1:M_out
    PSI_n[:,i] = Psi(X_rm_n[:,i])
end

for i = M_out + 1 : N
    X_rm_n[:,i] = sum(h_wf[:,:,k]*PSI_n[:,i-k] for k = 1:M_out, dims = 2) + sqrt(h)*σ*randn(nu,1)
    PSI_n[:,i] = Psi(X_rm_n[:,i])
end
```

```julia
plot(X_rm_n[1,:],X_rm_n[2,:])
```

```julia
data = Dict(
        "h_wf" => h_wf,
        "A" => A,
        "σ" => σ,
        "Xo" => Xo,
        "t_disc" => t_disc,
        "gap" => gap,
        "t_start" => t_start,
        "t_stop" => t_start,
        "h" => h,
        "X" => X,
        "X__rm" => X_rm,
        "X__rm_n" => X_rm_n)

save("Data\\LSDE_wfs.jld",data)
```
