```julia
using PyPlot
using StatsBase

include("modgen_linearSDE.jl")
```

```julia
A = -[0.1 10; 0 0.9]
σ = [1 0; 0 1]
Xo = [1; 1]
t_disc = 1000
gap = 1
scheme = "EM"

X = modgen_LSDE(0,10000,0.01,
    A = A,
    σ = σ,
    Xo = Xo,
    t_disc = t_disc,
    gap = gap,
    scheme = scheme)
```

```julia
plot(X[1,1:200:end],X[2,1:200:end])
```

```julia
A1 = autocor(X[1,:],1:4000)
A2 = autocor(X[2,:],1:4000)
plot(A1)
plot(A2)
```
