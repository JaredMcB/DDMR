# Scratch for the modReduction program

## Test Data 1

Here we generate the test data we will simply use Lorenz63

```julia
using PyPlot
using PyCall
so = pyimport("scipy.optimize")

sig = 10
bet = 8/3
rho = 28  

Steps = 10^4
Start = 0
Stop = 100
t = range(Start, length = Steps, stop = Stop)
h = t[2] - t[1]
x_init = [1;1;1]

F(x) = [sig*(x[2] - x[1]); 
    x[1]*(rho - x[3]) - x[2];  
    x[1]*x[2] - bet*x[3]]

X = zeros(3,Steps)
X[:,1] = x_init
for n = 1: Steps-1
    X[: , n+1] = X[:, n] + h*F(X[:,n]) + sqrt(h)*rand(3)
end

plot(t[1:1000],X[1, 1:1000],t[1:1000],X[2, 1:1000],t[1:1000],X[3, 1:1000])
```

## Test Data 2

Overdampled Langiven 

```julia
Start = 0
Stop = 10^4
Steps = Stop*100 + 1

F(x) = -x.*(x.^2 -[1]) 

x_init = [.5]

t = range(Start, length = Steps, stop = Stop)
h = t[2] - t[1]

X = zeros(1,Steps)
X[:,1] = x_init

for n = 1 : Steps - 1
    X[:,n+1] = X[:,n] + h*F(X[:,n]) + sqrt(h)*randn(1)
end

plot(t[1:1000],X[1,1:1000])
```

## The Model reduction function

We start by writting the auxillary function. This helps for the objective function. Here are the arguments:

```julia
p = 2
q = 2
r = floor(Int,p/2)
Psi(x) = [F(x) x]

d, N = size(X)
nu = size(Psi(X[:,1]),2)

A = randn(p + nu * (q +1)); A = [0 0 0 0 0 0 h 1]
```

```julia
function AUX(A,X,Psi,d,N,p,q,r,nu)
    beta = zeros(p-r)
    alpha = A[1 : p-r]
    beta[1:r] = A[p-r+1 : p]
    b = reshape(A[p+1 : end],nu,q+1)

    z = zeros(p-r,d,N)

    # Initialize the cascade
    z[end,:,1:p] = X[:,2 : p+1]

    for rr = p - r - 1 : -1 : 1
        for n = p : -1 : 3
            z[rr,:,n] = z[rr + 1, : , n] + alpha[rr+1]*z[rr + 1, : , n-1] + 
            beta[rr+1]*z[rr + 1, : , n-2]
        end
    end

    Y = zeros(d,Steps)
    Y[:,1:p+1] = X[:,1:p+1]

    # run the cascade
    for n = p+1 : N - 1
        Z = sum([Psi(X[:,n-p+j - 1])*b[:,j] for j = 1 : q+1])
        
        rr = 1
        z[rr,:,n] = -alpha[rr]*z[rr,:,n-1] - beta[rr] * (if p > 1 z[rr,:,n-2] else zeros(d) end) + Z    
        for rr = 2 : r
            z[rr,:,n] = -alpha[rr]*z[rr,:,n-1] - 
                beta[rr] * (if p > 1 z[rr,:,n-2] else zeros(d) end) + z[rr-1,:,n]
        end

        Y[:,n+1] = z[p - r,:,n] + sqrt(h)*randn(d)
       
    end
    return Y
end
```

```julia
Y = AUX(A,X,Psi,d,N,p,q,r)
Z = Y - X
plot(t[1:1000],X[1:1000])
plot(t[1:1000],Y[1:1000])
```

```julia
Y = AUX(A,X,Psi,d,N,p,q,r)
plot(t[1:1000],X[1, 1:1000],t[1:1000],X[2, 1:1000],t[1:1000],X[3, 1:1000])
plot(t[1:1000],Y[1, 1:1000],t[1:1000],Y[2, 1:1000],t[1:1000],Y[3, 1:1000])
```

```julia
function RUN(A,X,Psi,d,N,p,q,r,nu,h)
    beta = zeros(p-r)
    alpha = A[1 : p-r]
    beta[1:r] = A[p-r+1 : p]
    b = reshape(A[p+1 : end],nu,q+1)

    z = zeros(p-r,d,N)

    # Initialize the cascade
    z[end,:,1:p] = X[:,2 : p+1]

    for rr = p - r - 1 : -1 : 1
        for n = p : -1 : 3
            z[rr,:,n] = z[rr + 1, : , n] + alpha[rr+1]*z[rr + 1, : , n-1] + 
            beta[rr+1]*z[rr + 1, : , n-2]
        end
    end

    Y = zeros(d,Steps)
    Y[:,1:p+1] = X[:,1:p+1]

    # run the cascade
    for n = p+1 : N - 1
        Z = sum([Psi(Y[:,n-p+j - 1])*b[:,j] for j = 1 : q+1])
        
        rr = 1
        z[rr,:,n] = -alpha[rr]*z[rr,:,n-1] - beta[rr] * (if p > 1 z[rr,:,n-2] else zeros(d) end) + Z    
        for rr = 2 : r
            z[rr,:,n] = -alpha[rr]*z[rr,:,n-1] - 
                beta[rr] * (if p > 1 z[rr,:,n-2] else zeros(d) end) + z[rr-1,:,n]
        end

        Y[:,n+1] = z[p - r,:,n] + sqrt(h)*randn(d)
    end
    return Y
end
```

```julia
YY = RUN(A,X,Psi,d,N,p,q,r,nu,h)
```

```julia
plot(t[1:1000],X[1:1000])
plot(t[1:1000],YY[1:1000])
```
